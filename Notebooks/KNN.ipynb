```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/acidic_basic_classification.csv").drop("Sample_ID", axis = 1)
data
```

```python
data = data.sort_values("Classification").reset_index().drop("index", axis = 1)
data
```

```python
data0 = data[:7]
data1 = data[7:]
```

```python
plt.scatter(data0["pH"], data0["Temperature (°C)"])
plt.scatter(data1["pH"], data1["Temperature (°C)"])
```

```python
x = data[["pH", "Temperature (°C)", "Conductivity (mS/cm)"]]
y = data["Classification"]
```

```python
model = KNeighborsClassifier(n_neighbors = 3)

model.fit(x,y)
```
